```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import os
```

```python

basepath = "/work3/frwa/repos/bayesian-laplace-metric-learning/outputs/"
model_paths = [("contrastive", "Backbone/figures/FashionMNIST/train_script/epoch_100"),
               ("pfe", "PFE/figures/FashionMNIST/train_script/epoch_100"),
               #("posthoc", "PostHoc/figures/FashionMNIST/train_script/epoch_1"),
               ("posthoc_full", "PostHoc/figures/FashionMNIST/train_script_full/epoch_1"),
               ]


columns = ["test_accuracy", "test_map_k", "test_recall_k"]
additional_columns = ["test_ece", "test_ausc", "test_auroc", "test_auprc"]
uq_metrics = ["ausc"]
data = [["model"] + columns+additional_columns+uq_metrics]
for model, path in model_paths:
    
    # metrics
    with open(basepath + path + "/metrics.json") as f:
        metrics = json.load(f)
    
    row = [model]
    for column in columns:
        row.append(metrics[column])
    
    # additional metrics
    if os.path.exists(basepath + path + "/additional_metrics.json"):
        with open(basepath + path + "/additional_metrics.json") as f:
            hessian = json.load(f)
            
        for column in additional_columns:
            row.append(hessian[column])
    else:
        for column in additional_columns:
            row.append(np.nan)
            
            
    # metrics
    if os.path.exists(basepath + path + "/uncertainty_metrics.json"):
        with open(basepath + path + "/uncertainty_metrics.json") as f:
            metrics = json.load(f)
        
        for column in uq_metrics:
            row.append(metrics[column])
    else:
        for column in uq_metrics:
            row.append(np.nan)
        
    data.append(row)
        
data = pd.DataFrame(data)
```
